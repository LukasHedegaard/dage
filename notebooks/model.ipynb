```python
import tensorflow as tf
keras = tf.compat.v2.keras
import sys
sys.path.append('..')
import models
import losses
```

```python
INPUT_SHAPE=(32,32,1)
OUTPUT_SHAPE=(10)
optimizer=keras.optimizers.Adam()
aux_loss = losses.dnse_loss(margin=1)
model_base = models.common.conv2_block(input_shape=INPUT_SHAPE)
model, model_test = models.two_stream_pair_embeds.model(model_base=model_base, input_shape=INPUT_SHAPE, output_shape=OUTPUT_SHAPE, dense_size=120, embed_size=84, optimizer=optimizer, batch_size=16, aux_loss=aux_loss)
```

```python
keras.utils.plot_model( #type: ignore
    model,
    to_file='figures/model.png',
    show_shapes=True,
    show_layer_names=True,
    rankdir='TB',
    expand_nested=True,
    dpi=96
)
```
